```python
from __future__ import print_function
import cv2 as cv
import argparse

parser = argparse.ArgumentParser(description='Code for Histogram Equalization tutorial.')
parser.add_argument('--input', help='Path to input image.', default='anhcu.jpg')
args = parser.parse_args(args=[])

src = cv.imread(cv.samples.findFile(args.input))
if src is None:
    print('Could not open or find the image:', args.input)
    exit(0)

src = cv.cvtColor(src, cv.COLOR_BGR2GRAY)

dst = cv.equalizeHist(src)
```

```python
transformer.save(dst, "dst_thienbinh.jpg")
display(Image(filename='/Users/chibangnguyen/Documents/BigImage/image_transform/dst_thienbinh.jpg'))
```
